```python
# 例2.15
from matplotlib import pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

def Fun(x1,x2):
    return x1 - x2 + 2*x1**2 + 2*x1*x2 + x2**2

def Dx1Fun(x1,x2):
    return 1 - 4*x1 + 2*x2

def Dx2Fun(x1,x2):
    return -1 + 2*x1 + 2*x2

fig = plt.figure()
#ax = Axes3D(fig)
ax = fig.add_subplot(111, projection='3d')
ax.view_init(elev=30, azim=90)
X1,X2 = np.mgrid[-2:2:40j,-2:2:40j]
Z = Fun(X1,X2)
ax.plot_surface(X1,X2,Z,rstride=1,cstride=1,cmap='rainbow')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('z')

step = 0.01
x1 = 0
x2 = 0
z = Fun(x1,x2)
tag_x1 = [x1]
tag_x2 = [x2]
tag_z = [z]
over = False
for i in range(1000):
    x1_new = x1 - step*Dx1Fun(x1,x2)
    x2_new = x2 - step*Dx2Fun(x1,x2)
    z_new = Fun(x1_new,x2_new)
    if abs(z_new - z) < 1e-6:
        over = True
    else:
        x1 = x1_new
        x2 = x2_new
        z = z_new
        tag_x1.append(x1)
        tag_x2.append(x2)
        tag_z.append(z)

ax.plot(tag_x1,tag_x2,tag_z,color='red')

plt.show()
```
