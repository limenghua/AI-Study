```python
from nltk.stem.porter import PorterStemmer
from nltk.stem.lancaster import LancasterStemmer
from nltk.stem.snowball import SnowballStemmer

input_words = {'reading', 'calves','acting','undefined','house','possibly','version','hotel','learned','experience'}

porter = PorterStemmer()
lancaster = LancasterStemmer()
snowball = SnowballStemmer('english')

stemmer_names = ['Porter', 'Lancaster', 'Snowball']

formatted_text = '{:>16}' * (len(stemmer_names) + 1)
print('\n',formatted_text.format('Input Word', *stemmer_names),'\n','*' * 70)
for word in input_words:
    output = [word,porter.stem(word),lancaster.stem(word),snowball.stem(word)]
    print(formatted_text.format(*output))
```

```python
from nltk.stem.porter import PorterStemmer
from nltk.stem.lancaster import LancasterStemmer
from nltk.stem.snowball import SnowballStemmer
from nltk.stem.wordnet import WordNetLemmatizer

input_words = {'reading', 'calves','acting','undefined','house','possibly','version','hotel','learned','experience'}

porter = PorterStemmer()
lancaster = LancasterStemmer()
snowball = SnowballStemmer('english')
wordnet = WordNetLemmatizer()

stemmer_names = ['Porter', 'Lancaster', 'Snowball','Wordnet Noun','Wordnet Verb']

formatted_text = '{:>20}' * (len(stemmer_names) + 1)
print('\n',formatted_text.format('Input Word', *stemmer_names),'\n','*' * 120)
for word in input_words:
    porter_stem = porter.stem(word)
    lancaster_stem = lancaster.stem(word)
    snowball_stem = snowball.stem(word)
    wordnet_noun_stem = wordnet.lemmatize(word, pos='n')
    wordnet_verb_stem = wordnet.lemmatize(word, pos='v')
    output = [word,porter_stem,lancaster_stem,snowball_stem,wordnet_noun_stem,wordnet_verb_stem]
    print(formatted_text.format(*output))
```
