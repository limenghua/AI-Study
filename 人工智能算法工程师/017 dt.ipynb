```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn import tree

import matplotlib.pyplot as plt
```

```python
iris = load_iris()
print(iris.feature_names)
print(iris.target_names)
```

```python
# 划分数据集
removed = [0,50,100]
new_target = np.delete(iris.target, removed)
new_data = np.delete(iris.data, removed, axis=0)

# 训练分类器
clf = tree.DecisionTreeClassifier()
clf.fit(new_data, new_target)
prediction = clf.predict(iris.data[removed])

print("Original Label:", iris.target[removed])
print("Label Predicted:",prediction)
```

```python
# 绘制决策树
plt.figure(figsize=(15, 10))
tree.plot_tree(clf, feature_names=iris.feature_names, filled=True)
plt.show()
```
