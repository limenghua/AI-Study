# 使用最小二乘法求解线性回归


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class LinearRegression:
    def __init__(self):
        self.theta_ = None


    def fit(self, X, y):
        X = np.hstack((np.ones((X.shape[0], 1)), X))
        self.theta_ = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)

    def predict(self, X):
        X = np.hstack((np.ones((X.shape[0], 1)), X))
        return X.dot(self.theta_)
```

```python
#load data
import pandas as pd
data=pd.read_csv('data/data1.csv')
# 数据赋值
x=data.loc[:,'x']
y=data.loc[:,'y']

# 画图
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
model = LinearRegression()
x = np.array(x).reshape(-1, 1)
model.fit(x,y)
```

```python
y_pred = model.predict(x)
# 计算R^2
from sklearn.metrics import r2_score
r2 = r2_score(y, y_pred)
print("R^2:", r2)
```

```python
# 预测x = 3.5
y_pred = model.predict(x)

fig1 = plt.figure(figsize=(8, 6))
plt.scatter(x,y)
plt.plot(x,y_pred,color='red',linewidth=1)
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```
