```python
# 密度聚类 算法的代码实现

import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
def load_data():
    data = np.loadtxt('data/cluster_data.csv',delimiter=',')
    return data
```

```python
X = load_data()

fig = plt.figure(figsize=(6, 5))
plt.scatter(X[:, 0], X[:, 1], cmap='cm_dark', s=20)
plt.title('Original Data')
plt.show()
```

```python
from sklearn.cluster import DBSCAN

model = DBSCAN(eps=0.5, min_samples=5,metric='euclidean')
model.fit(X)
```

```python

print("每个样本所属的类别:",model.labels_)
```

```python
labels = model.labels_
# 结果可视化

fig4 = plt.figure(figsize=(6, 5))
cm_dark = mpl.colors.ListedColormap(['g','r','b','c'])
plt.scatter(X[:, 0], X[:, 1], c=labels, s=20, cmap=cm_dark)
plt.title('DBSCAN Clustering')
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```
