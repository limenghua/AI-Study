```python
# K-means 算法的代码实现

import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_data():
    data = np.loadtxt('data/cluster_data.csv',delimiter=',')
    return data
```

```python
X = load_data()

fig = plt.figure(figsize=(6, 5))
plt.scatter(X[:, 0], X[:, 1], cmap='cm_dark', s=20)
plt.title('Original Data')
plt.show()
```

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=3,max_iter=100)
model.fit(X)
```

```python
print("聚类中心:",model.cluster_centers_)
print("每个样本所属的类别:",model.labels_)
```

```python
centroids = model.cluster_centers_
labels = model.labels_
# 结果可视化

fig4 = plt.figure(figsize=(6, 5))
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='*', s=200, c='red')
plt.title('K-means Clustering')
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```
