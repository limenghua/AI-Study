```python
import numpy as np
import matplotlib.pyplot as plt
```

梯度下降法
求y = $x^2$ + 2x + 5 的最小值

y' = 2x + 2


```python
#可视化函数图像
x = np.linspace(-10,10,100)
y = x**2 + 2*x + 5
plt.plot(x,y)
plt.show()
```

```python
x=3
alpha=0.1
n=100

for i in range(n):
    x = x - alpha*(2*x+2)

y= x**2 + 2*x + 5
print(x,y)
```
