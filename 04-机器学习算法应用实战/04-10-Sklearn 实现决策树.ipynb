```python
from sklearn.tree import DecisionTreeClassifier
import numpy as np

from sklearn.tree import export_graphviz
```

```python
def loaddata():
    dataSet = [[0,0,0,0,'no'],
               [0,0,0,1,'no'],
               [0,1,0,1,'yes'],
               [0,1,1,0,'yes'],
               [0,0,0,0,'no'],
               [1,0,0,0,'no'],
               [1,0,0,1,'no'],
               [1,1,1,1,'yes'],
               [1,0,1,2,'yes'],
               [1,0,1,2,'yes'],
               [2,0,1,2,'yes'],
               [2,0,1,1,'yes'],
               [2,1,0,1,'yes'],
               [2,1,0,2,'yes'],
               [2,0,0,0,'no']]
    featureNames = ['age','job','hosuse','credit','class']
    return dataSet, featureNames
```

```python
dataSet,featureNames = loaddata()
X = np.array(dataSet)[:,:-1]
y = np.array(dataSet)[:,-1]

# 决策树分类器
clf = DecisionTreeClassifier(criterion='entropy', min_samples_split=5)
clf.fit(X, y)

# 预测
print(clf.predict([[2, 1,1,1]]))
```

```python
from sklearn import tree
tree.plot_tree(clf, filled=True, 
feature_names=featureNames, class_names=['yes','no'])
```
