# 06-16-Sklearn 实现SVM2 输出常用属性值

```python
# 使用numpy 读取数据集
import numpy as np

def load_data(file_name):
    data = np.loadtxt(file_name, delimiter='\t')
    X = data[:, :-1]
    y = data[:, -1]
    return X, y
```

```python
# load data
X,y = load_data('data/svm1.txt')
```

```python
# 建立SVM 分类器并输出常用属性值
from sklearn.svm import SVC
model = SVC(kernel='linear', C=1.0)
model.fit(X, y)
print("截距项：", model.intercept_)
print("权重系数：", model.coef_)
print("支持向量的个数：", model.n_support_)
print("支持向量在样本中的索引：", model.support_)
print("支持向量对应的标签：", model.support_vectors_)

print("支持向量的alpha值：", model.dual_coef_)
print("支持向量机的评估分数：", model.score(X, y))
```

```python
w= model.coef_
b= model.intercept_
print("w:", w)
print("b:", b)
```

```python
# 数据可视化
# 绘制分割超平面
import matplotlib.pyplot as plt
fig2 = plt.figure()
plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=plt.cm.Paired)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

# 绘制决策平面
x0 = np.linspace(X[:, 0].min(), X[:, 0].max(), 100)
x1 = (-w[0][0] * x0 - b) / w[0][1]

plt.plot(x0, x1, 'k-', label='Decision boundary')

# 绘制支持向量
sv = model.support_vectors_
plt.scatter(sv[:, 0], sv[:, 1], s=100, facecolors='none', edgecolors='k', label='Support vectors')


plt.legend()
plt.show()
```
